# Homework code for week 1

This notebook shows the code used for generating the figures required for the homework of week 1.
The intention was to do this in SPSS, but as I am using Linux, and GNU PSPP is not good enough for it, I decided to use Python 3 with Matplotlib, Numpy, and Scipy.
This notebook was evaluated in Python 3.7, but should work in Python 3.6.
It will not work in Python 3.5 or lower because type annotations and the `typing` module are used.

Much of the code required to evaluate this notebook is placed in the companion file named `utils.py` so that it is reusable.

```python
# Import all necessary functions from `utils.py` in order to show all figures.
from utils import summary, histogram, boxplot, qq_norm, qq_exp

# The name to show in the headers (or on the axes) of the plots and tables.
name = "Playtimes_Rolf"

# The data taken from the table on the excersises.
measurements = [
     2.0,  2.2,  2.9,  3.0,  3.7,  3.9,  4.0,  4.1,  5.8,  5.9,
     6.3,  6.8,  6.9,  7.2,  7.4,  7.5,  8.4,  9.0,  9.2,  9.2,
     9.4,  9.8,  9.9, 10.0, 10.5, 11.1, 11.2, 11.3, 11.3, 11.5,
    12.0, 13.0, 13.2, 15.8, 19.6, 22.7,
]

# Instructs matplotlib to output images instead of attempting to open an X-window.
%matplotlib inline

# Draws the summary of the dataset (mean, standard deviation, etcetera).
summary([measurements], [name])

# Draws a histogram (normalized to surface) with a normal distribution.
histogram(measurements, name)

# Draws a boxplot showing the four quartiles and all the outliers.
boxplot([measurements], [name])

# Draws a Q-Q Plot fitting the dataset to a normal (scipy `norm`) distribution
qq_norm(measurements, name)

# Draws a Q-Q Plot fitting the dataset to an exponential (scipy `expon`) distribution.
qq_exp(measurements, name)
```
